```python
import plotly.graph_objects as go
import yfinance
```

```python
#Retrieve Amazon stock for the last five years
amzn = yfinance.Ticker('AMZN')
hist = amzn.history(period='5y')
```

```python
#Display Amazon data
hist.head()
```

```python
# Disply last five years of Amazon stocks
amzn_fig = go.Figure(data=go.Scatter(x=hist.index,y=hist['Close'], mode='lines'))
amzn_fig.show()
```
