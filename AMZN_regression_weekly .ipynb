```python
# The train_test_split function is for splitting a single dataset for two different
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
from yahoo_fin.stock_info import get_data
import pandas as pd
```

```python
df= get_data("amzn", start_date="01/01/2020", end_date="12/31/2020", index_as_date = False, interval="1wk")
df
```

```python
df1= get_data("amzn", start_date="01/01/2016", end_date="12/31/2019", index_as_date = False, interval="1wk")
df1
```

```python
df1= df1.drop('ticker',axis=1)
df1
```

```python
df1 = df1[['date', 'close']]

df1
```

```python
df.index = df.pop('date')
df
```

```python
# creat a variable to Predict 2020 days 
prediction_days= 53
# Create a new column target data - Predict values x days away from current price
df1['Actual Prediction'] = df1[['close']].shift(-prediction_days)
df1
```

```python
# Create independent dataset X and create a numpy array and 
X=X = np.array(df1[['close']])

#remove the NaN days/rows
X=X[:-prediction_days]
print(X)
```

```python
#Creat the dependetnt dadaset Y
Y=df1['Actual Prediction'].values
Y=Y[:-prediction_days]
print(Y)
```

```python
# train and test ratio is 75% traing and 25% testing
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
#Creat the model
linr = LinearRegression()
#train the model
linr.fit(x_train, y_train)
```

```python
#test the model using score
linr_score=linr.score(x_test,y_test)
print(linr_score)
```

```python
#predict the last 52 days(2020 days)

#1-varible has the actual close value of the the last 52 days
acutal_52days=np.array(df1[['close']])[-prediction_days:]
acutal_52days
```

```python
#print the liner regression prediction  for the next 52 days

linr_prediction=linr.predict(acutal_52days)

linr_prediction
```
